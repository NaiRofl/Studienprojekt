```python
# aggregiert die Emissionen aus json-Dateien für jedes Bundesland und jedes Land, berechnet die Gesamtemissionen in Kilogramm und 
# speichert diese in einem dict. Die Top-N Funktion sortiert Länder nach Gesamtemissione.

import os
import json

folder_path = r'C:\Users\flori\OneDrive\Desktop\Studium\Python\08_Studienprojekt\Emission-Data\Evaluate_Emissions_Country\2022'
gesamtemissionen_bundesland_DE = {}


for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        file_path = os.path.join(folder_path, filename)
        
        with open(file_path, 'r') as file:
            data = json.load(file)
        
        bundesland = filename.replace('.json', '')

        if bundesland not in gesamtemissionen_bundesland_DE:
            gesamtemissionen_bundesland_DE[bundesland] = {}
            print(bundesland)

        for land, emissionen in data.items():
            if land not in gesamtemissionen_bundesland_DE[bundesland]:
                gesamtemissionen_bundesland_DE[bundesland][land] = 0

            gesamt_emissionen_land = sum(emissionen.values()) / 1000
            gesamtemissionen_bundesland_DE[bundesland][land] += gesamt_emissionen_land


def top_n_emissionen(emissionen_dict, n=10):
    sortierte_emissionen = sorted(emissionen_dict.items(), key=lambda x: x[1], reverse=True)
    top_n = sortierte_emissionen[:n]
    return top_n
```

```python
# berechnet die monatlichen und jährlichen Gesamtemissionen für ein bestimmtes Bundesland aus json-Dateien, indem er die 
# Emissionen für jeden Monat summiert und die Gesamtemissionen ausgibt

import json
import os

def sum_over_time(start_date: str, end_date: str, data: dict) -> float:
    counter = 0
    count = False
    for date, value in data.items():
        if date == start_date:
            count = True
        if count:
            counter += value
        if date == end_date:
            break
    return counter

folder_path = r'C:\Users\flori\OneDrive\Desktop\Studium\Python\08_Studienprojekt\Emission-Data\Evaluate_Emissions_Timestamp_DE\2023'

monate_2023 = [
    ("Januar", "2023-01-01 00:00:00", "2023-01-31 23:00:00"),
    ("Februar", "2023-02-01 00:00:00", "2023-02-28 23:00:00"),
    ("März", "2023-03-01 00:00:00", "2023-03-31 23:00:00"),
    ("April", "2023-04-01 00:00:00", "2023-04-30 23:00:00"),
    ("Mai", "2023-05-01 00:00:00", "2023-05-31 23:00:00"),
    ("Juni", "2023-06-01 00:00:00", "2023-06-30 23:00:00"),
    ("Juli", "2023-07-01 00:00:00", "2023-07-31 23:00:00"),
    ("August", "2023-08-01 00:00:00", "2023-08-31 23:00:00"),
    ("September", "2023-09-01 00:00:00", "2023-09-30 23:00:00"),
    ("Oktober", "2023-10-01 00:00:00", "2023-10-31 23:00:00"),
    ("November", "2023-11-01 00:00:00", "2023-11-30 23:00:00"),
    ("Dezember", "2023-12-01 00:00:00", "2023-12-31 23:00:00")
]

def emissions_for_bundesland(bundesland: str):
    emissionen_pro_monat = {}

    for filename in os.listdir(folder_path):
        if filename.endswith('.json') and filename.replace('.json', '') == bundesland:
            file_path = os.path.join(folder_path, filename)
            # print(filename)
            

            with open(file_path, 'r') as file:
                data = json.load(file)
            

            for monat, start_date, end_date in monate_2023:
                emissionen = sum_over_time(start_date, end_date, data)
                if monat not in emissionen_pro_monat:
                    emissionen_pro_monat[monat] = 0
                emissionen_pro_monat[monat] += emissionen


    sum_of = 0
    for monat, emissionen in emissionen_pro_monat.items():
        sum_of += emissionen
        print(f"{monat}: {emissionen / 1000:.2f} tCO2")

    return sum_of / 1000


bundesland = '2023_import_table_total_DE_NW.csv_emissions_per_timestamp'
total_emissions = emissions_for_bundesland(bundesland)
print(f"Gesamtemissionen für {bundesland}: {total_emissions:.2f} tCO2")
```
